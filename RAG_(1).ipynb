```python
# @title Default title text
import os

image_metadata_dict = {}
base_path = '/content/drive/MyDrive/data/images'
image_folders = os.listdir(base_path)
one_image_paths = ['Loài nguy cấp', 'Loài thông thường']

for image_folder in image_folders:
    folder_path = os.path.join(base_path, image_folder)

    for one_image_path in one_image_paths:
        image_path = os.path.join(folder_path, one_image_path)

        if not os.path.isdir(image_path):
            continue

        for image_file in os.listdir(image_path):
            img_path = os.path.join(image_path, image_file)

            if img_path.endswith(('.jpg', '.webp', '.jpeg', 'jfif', '.png', '.gif')):
                image_metadata_dict[img_path] = {
                    'filename': image_file,
                    'img_path': img_path
                }

            elif os.path.isdir(img_path):  # Kiểm tra nếu là thư mục
                for img_file in os.listdir(img_path):
                    full_img_path = os.path.join(img_path, img_file)
                    if full_img_path.endswith(('.jpg', '.webp', 'jfif', '.jpeg', '.png', '.gif')):
                        image_metadata_dict[full_img_path] = {
                            'filename': img_file,
                            'img_path': full_img_path
                        }
```

```python
import spacy
import re

seq_len = 400
page_and_texts = []

def text_formatter(text: str) -> str:
    cleaned_text = text.replace("\n", " ").strip()
    cleaned_text = cleaned_text.replace("÷", " đến ")
    cleaned_text = re.sub(r'\.+', '.', cleaned_text)
    # cleaned_text = cleaned_text.replace(":", " ")
    cleaned_text = cleaned_text.replace("<", "nhỏ hơn")
    cleaned_text = cleaned_text.replace("≥", "lớn hơn bằng")
    cleaned_text = cleaned_text.replace(">", "lớn hơn")
    cleaned_text = cleaned_text.replace("≤", "nhỏ hơn bằng")
    return cleaned_text

def split_chunk(input_list: list) -> list[list[str]]:
    slice_size = 2
    list_sentence = [input_list[i:i + slice_size] for i in range(0, len(input_list), slice_size)]
    chunks = []
    for chunk in list_sentence:
        chunk = "".join(chunk).replace("  ", " ").strip()
        if len(chunk) > seq_len:
            nlp = spacy.blank("vi")
            nlp.add_pipe("sentencizer")
            chunk = list(nlp(chunk).sents)
            chunk = [str(sentence) for sentence in chunk]
            chunks.extend(chunk)
        else:
            chunks.append(chunk)
    return chunks

def sentences_split(texts):
    nlp = spacy.blank("vi")
    nlp.add_pipe("sentencizer")
    texts = list(nlp(texts).sents)
    texts = [str(sentence) for sentence in texts]
    for sentence in texts:
        if len(sentence) > seq_len:
            texts.remove(sentence)
            sentence = sentence.replace(";", ".")
            sentence = list(nlp(sentence).sents)
            for item in sentence:
                texts.append(str(item))
    texts = split_chunk(texts)
    return texts

def split_section(text: str) -> list[dict]:
    sections = [{'title': "", 'content': ""}]
    cur_main_sec = ""
    for line in text.split('\n'):
        title_match = re.match(r"(Tên:)", line)
        if not title_match:
            if cur_main_sec == "":
                sections[-1]['content'] += line + " "
            else:
                sections[-1]['content'] += " " + line
        else:
            cur_main_sec = line
            sections.append({'title': cur_main_sec, 'content': ""})
    return sections
for document in documents:
    text = document.text
    sections = split_section(text)
    for section in sections:
        cleaned_content = text_formatter(section['content'])
        len_content = len(cleaned_content)
        if len_content > seq_len:
            cleaned_content = sentences_split(cleaned_content)
        else:
            cleaned_content = [cleaned_content]
        for i in range (len(cleaned_content)):
          page_and_texts.append({'text': f"{section['title']}\n {cleaned_content[i]}"})
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def plot_images(image_metadata_dict):
  images = []
  image_shown = 0
  for image_filename in image_metadata_dict:
    img_path = image_metadata_dict[image_filename]['img_path']
    if os.path.isfile(img_path):
      filename = image_metadata_dict[image_filename]['filename']
      image = Image.open(img_path).convert('RGB')

      plt.subplot(8,8,len(images) + 1)
      plt.imshow(image)
      plt.xticks([])
      plt.yticks([])

      images.append(filename)
      image_shown += 1
      if image_shown >= 64:
        break
  plt.tight_layout()
plot_images(image_metadata_dict)
```

```python
import torch
from transformers import AutoProcessor, CLIPModel
from llama_index.core import (
    ServiceContext,
    SimpleDirectoryReader,
    VectorStoreIndex, StorageContext
)
from llama_index.vector_stores.chroma import ChromaVectorStore
import chromadb
import numpy as np

# Load mô hình CLIP local sentence-transformers/clip-ViT-B-32-multilingual-v1
model_id = "sentence-transformers/clip-ViT-B-32-multilingual-v1"
processor = AutoProcessor.from_pretrained(model_id)
model = CLIPModel.from_pretrained(model_id)

# Chuyển model sang GPU nếu có
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# Hàm nhúng văn bản bằng CLIP
def clip_embed_text(texts):
    inputs = processor(text=texts, return_tensors="pt", padding=True, truncation=True, max_length=77)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        text_features = model.get_text_features(**inputs)  # Đảm bảo model hỗ trợ hàm này
    return text_features.cpu().numpy()

texts = [item['text'] for item in page_and_texts]

# Chuyển văn bản thành vector embeddings bằng CLIP
text_embeddings = clip_embed_text(texts)

# Khởi tạo ChromaDB client (lưu vào thư mục "chromadb_data")
chroma_client = chromadb.PersistentClient(path="./chromadb_data")

# Tạo collection trong ChromaDB
collection = chroma_client.get_or_create_collection("birds_text")

# Lưu vector embeddings vào ChromaDB
for i, (text, vector) in enumerate(zip(texts, text_embeddings)):
    collection.add(
        ids=[str(i)],  # ID duy nhất
        embeddings=[vector.tolist()],  # Chuyển numpy array thành danh sách
        metadatas=[{"text": text}]
    )

print("Dữ liệu đã lưu vào ChromaDB!")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, pipeline
import torch

model_name = "deepseek-ai/deepseek-llm-7b-chat"

# Cấu hình tải mô hình với 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)

# Tải mô hình và tokenizer
model_deepseek = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Định nghĩa generator với pipeline
generator = pipeline("text-generation", model=model_deepseek, tokenizer=tokenizer)

```

```python

def query_chroma(text_query,output_language='vi'):
    # 1. Tìm kiếm dữ liệu liên quan trong ChromaDB
    query_embedding = clip_embed_text([text_query])
    results = collection.query(query_embeddings=[query_embedding.tolist()[0]], n_results=1)
    # Check if results["metadatas"] is not empty and has an element at index 0
    if results["metadatas"] and results["metadatas"][0]:
        retrieved_docs = [meta["text"] for meta in results["metadatas"][0]]
    else:
        retrieved_docs = []  # Handle empty results by returning an empty list

    # 2. Kết hợp dữ liệu tìm được với câu hỏi
    context = "\n".join(retrieved_docs)
    # Thêm hướng dẫn ngôn ngữ vào truy vấn
    full_query = f"Output in: {output_language}"

    # 3. Gửi truy vấn vào DeepSeek (sử dụng pipeline)
    deepseek_response = generator(full_query, max_new_tokens=100)[0]  # Use max_new_tokens to control generated length # Điều chỉnh max_length nếu cần
    return [{"text": deepseek_response["generated_text"]}]
```

```python
import torch
from transformers import CLIPProcessor, CLIPModel
import numpy as np

# Load the CLIP model with ViT-B/32
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# The resolution of the input images expected by the model
input_resolution = model.vision_model.config.image_size

# The maximum length of the input text
context_length = model.text_model.config.max_position_embeddings

# The size of the vocabulary used by the model
vocab_size = model.text_model.config.vocab_size

# Print the information about the model to the console
print(
    "Model parameters:",
    f"{np.sum([int(np.prod(p.shape)) for p in model.parameters()]):,}",
)
print("Input resolution:", input_resolution)
print("Context length:", context_length)
print("Vocab size:", vocab_size)
```

```python
#set the device
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(device)

# img_emb_dict stores image embeddings for each image
img_emb_dict = {}
with torch.no_grad():
    # Iterate over the image metadata dictionary and extract image embeddings for each image
    for image_filename in image_metadata_dict:
        img_file_path = image_metadata_dict[image_filename]["img_path"]
        if os.path.isfile(img_file_path):
            image = Image.open(img_file_path)
            inputs = processor(images=image, return_tensors="pt").to(device)

            # Extract image features using the CLIP model
            image_features = model.get_image_features(**inputs)

            # Store the image features in the image embedding dictionary
            img_emb_dict[image_filename] = image_features
```

```python
from llama_index.core.schema import ImageDocument

# create a list of ImageDocument objects, one for each image in the dataset
img_documents = []
for image_filename in image_metadata_dict:
    # the img_emb_dict dictionary contains the image embeddings
    if image_filename in img_emb_dict:
        filename = image_metadata_dict[image_filename]["filename"]
        filepath = image_metadata_dict[image_filename]["img_path"]
        print(filepath)

        # create an ImageDocument for each image
        newImgDoc = ImageDocument(
            text=filename, metadata={"filepath": filepath}
        )

        # set image embedding on the ImageDocument
        newImgDoc.embedding = img_emb_dict[image_filename].tolist()[0]
        img_documents.append(newImgDoc)

# Khởi tạo ChromaDB client (lưu vào thư mục "chromadb_data")
chroma_client = chromadb.PersistentClient(path="./chromadb_data")

# Tạo collection trong ChromaDB cho ảnh
image_collection = chroma_client.get_or_create_collection("birds_img")

# Lưu vector embeddings vào ChromaDB
for i, (image_filename, embedding) in enumerate(img_emb_dict.items()):
    if embedding is not None:
        metadata = image_metadata_dict.get(image_filename, {})
        filename = metadata.get("filename", "unknown")
        filepath = metadata.get("img_path", "")

        # Chuyển embedding về dạng danh sách một chiều
        embedding_array = embedding if isinstance(embedding, list) else embedding.tolist()
        if isinstance(embedding_array[0], list):  # Nếu là danh sách lồng nhau, lấy phần tử đầu tiên
            embedding_array = embedding_array[0]

        image_collection.add(
            ids=[str(i)],  # ID duy nhất
            embeddings=[embedding_array],  # Chuyển numpy array thành danh sách một chiều
            metadatas=[{"filename": filename, "filepath": filepath}]
        )

print("Dữ liệu ảnh đã lưu vào ChromaDB!")
```

```python
def image_search(image_path, top_k=1):  # top_k là số lượng kết quả muốn trả về
    # Load hình ảnh và tiền xử lý (nếu cần)
    image = Image.open(image_path)
    inputs = processor(images=image, return_tensors="pt").to(device)

    # Trích xuất embeddings của hình ảnh
    with torch.no_grad():
        image_embedding = model.get_image_features(**inputs).cpu().numpy().tolist()[0]

    # Truy vấn ChromaDB
    search_results = image_collection.query(
        query_embeddings=[image_embedding],
        n_results=top_k
    )

    # Trả về kết quả tìm kiếm (metadata và khoảng cách)
    return search_results

def plot_image_search_results(image_search_results):
    plt.figure(figsize=(16, 5))

    img_cnt = 0
    for i in range(len(image_search_results["ids"][0])):
        img_name = image_search_results["metadatas"][0][i]["filename"]
        img_path = image_search_results["metadatas"][0][i]["filepath"]
        image = Image.open(img_path).convert("RGB")
        score = image_search_results["distances"][0][i]

        plt.subplot(2, 3, img_cnt + 1)
        plt.title("{:.4f}".format(score))

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        img_cnt += 1
```

```python
def get_image_embedding(image_path):
    # Load hình ảnh và tiền xử lý
    image = Image.open(image_path)
    inputs = processor(images=image, return_tensors="pt").to(device)

    # Trích xuất embeddings
    with torch.no_grad():
        image_embedding = model.get_image_features(**inputs).cpu().numpy().tolist()[0]

    return image_embedding
```

```python
def combined_search(image_path=None, text_query=None,output_language='vi', top_k=1):
    """
    Thực hiện tìm kiếm kết hợp bằng hình ảnh và text.

    Args:
        image_path: Đường dẫn đến hình ảnh (nếu có).
        text_query: Câu truy vấn text (nếu có).
        top_k: Số lượng kết quả muốn trả về.

    Returns:
        Danh sách kết quả tìm kiếm (metadata và khoảng cách).
    """

    if image_path:
        # Lấy embedding của hình ảnh
        image_embedding = get_image_embedding(image_path)

        # Truy vấn ChromaDB (collection "birds_img")
        image_search_results = image_collection.query(
            query_embeddings=[image_embedding],
            n_results=top_k
        )

        # Lấy metadata của hình ảnh được tìm thấy
        image_metadata = image_search_results['metadatas'][0][0]  # Lấy metadata của hình ảnh đầu tiên

        # Tạo câu truy vấn mới dựa trên metadata và text_query
        if text_query:
            new_query = f"{text_query} {image_metadata.get('filename', '')}"  # Kết hợp text_query và filename
        else:
            new_query = image_metadata.get('filename', '')  # Chỉ sử dụng filename nếu không có text_query
    else:
        image_search_results = None
        new_query = text_query # Sử dụng text_query gốc nếu không có image_path

    # Truy vấn ChromaDB (collection "birds_text") để lấy context
    text_search_results = query_chroma(new_query)

    return image_search_results, text_search_results
```

```python
# Sử dụng hàm combined_search
image_path = "/content/drive/MyDrive/data/images/1. Họ Bồ Nông/Loài nguy cấp/1. Bồ nông chân xám - Pelecanus philippensis/1. Bồ nông chân xám - Pelecanus philippensis (1).jpg"
text_query = "Đây là loài chim gì?"
image_results, text_results = combined_search(image_path, text_query, top_k=1)

# Hiển thị kết quả
if image_results:
    plot_image_search_results(image_results)  # Giả sử plot_image_search_results đã được định nghĩa
if text_results:
    print("Kết quả tìm kiếm text:")
    for result in text_results:
        print(result)
```
